```r
XOM_data <- read_csv("XOM.csv")
```

```r
XOM_data <- XOM_data |>
                mutate(year = substr(date, 1,4), month = substr(date, 6,7)) |>
                mutate(year = as.integer(year), month = as.integer(month))
                
```

```r
XOM_data
```

```r
Grouped_XOM <- XOM_data |>
                    group_by(year,month) |>
                    summarize(monthly_price_avg = mean(close, na.rm = TRUE),
                              monthly_vol_avg = mean(volume, na.rm = TRUE)) 
```

```r
Grouped_XOM
```

```r
previous_month_price <- Grouped_XOM |>
                            pull(monthly_price_avg)
```

```r
previous_month_price <- previous_month_price[-c(1)]

previous_month_price
```

```r
previous_month_vol <- Grouped_XOM |>
                            pull(monthly_vol_avg)
```

```r
previous_month_vol <- previous_month_vol[-c(1)]

previous_month_vol
```

```r
ADDED_XOM <- Grouped_XOM |>
                cbin(previous_month_price, previous_month_vol) 
```

```r
Grouped_XOM$previous_month_price <- previous_month_price
```
